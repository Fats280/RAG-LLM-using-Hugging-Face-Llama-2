```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, ServiceContext
from llama_index.core.prompts.prompts import SimpleInputPrompt
```

```python
!pip install llama-index-llms-huggingface
from llama_index.llms.huggingface import HuggingFaceLLM
```

```python
documents = SimpleDirectoryReader("/content/data").load_data()
```

```python
documents
```

```python
system_prompt="""
You are a Q&A assistant. Your goal is to answer questions as accurately
as possible based on the instructions and context provided.
"""
## Using a default format which is supported by Llama 2
query_wrapper_prompt = SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
import torch
from transformers import BitsAndBytesConfig

# quantize to save memory
quantization_config = BitsAndBytesConfig(
    load_in_8bit=True,
    bnb_8bit_compute_dtype=torch.float16,
    bnb_8bit_quant_type="nf8",
    bnb_8bit_use_double_quant=True,
    )

llm = HuggingFaceLLM(
     context_window=4096,
     max_new_tokens=256,
     generate_kwargs={"temperature":0.0, "do_sample": False},
     system_prompt=system_prompt,
     query_wrapper_prompt=query_wrapper_prompt,
     tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
     model_name="meta-llama/Llama-2-7b-chat-hf",
     device_map="auto",
     #model_kwargs={"quantization_config": quantization_config}
     )
```

```python
!pip install llama-index-embeddings-langchain
from llama_index.embeddings.langchain import LangchainEmbedding
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings

embed_model=LangchainEmbedding(HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2"))
```

```python
service_context=ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embed_model
)
```

```python
service_context
```

```python
index=VectorStoreIndex.from_documents(documents,service_context=service_context)
```

```python
index
```

```python
query_engine=index.as_query_engine()
```

```python
response = query_engine.query("Explain how BWA-MEM works.")
```

```python
print(response)  # final RAG-LLM output
```
